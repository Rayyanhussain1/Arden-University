# **Linear Algebra Analysis of the Population Interaction Model**
### **Model Explanation**
The relationship between the two groups is represented with a transition matrix, where the population in each year is calculated from the previous year’s values.

$$x_{n+1} = \begin{bmatrix} 1.2 & 0.1 \\ 0.05 & 1.1 \end{bmatrix}$$

This implies that:

Group X grows by about 20% per year and is also slightly affected by the size of Group Y.

Group Y grows by about 10% per year and is similarly influenced to a small extent by Group X.

### **Figure Interpretation**
From the graph, both populations rise rapidly over time in an exponential pattern. This happens because the largest eigenvalue of the transition matrix is above 1, which signals that the system does not settle to a fixed level but instead continues expanding.

```python
# Load required packages
import numpy as np
import matplotlib.pyplot as plt

# Define the growth/interaction matrix
transition_matrix = np.array([
    [1.2, 0.1],
    [0.05, 1.1]
])

# Starting values for the two groups
state = np.array([100, 80])

num_steps = 20
history = []

# Simulate population changes over time
for step in range(num_steps):
    history.append(state.copy())
    state = transition_matrix @ state

history = np.array(history)

# Draw the results
plt.plot(history[:, 0], label="Group X")
plt.plot(history[:, 1], label="Group Y")
plt.xlabel("Time (Years)")
plt.ylabel("Population Size")
plt.title("Projected Growth of Two Population Groups")
plt.legend()
plt.show()
```

# **Calculus Analysis of Population Growth in Paris**
Paris’s population is represented using an exponential model, where
𝑡
t measures time in years.

$$P(t) = 1,000e^{0.03t}$$


### **Growth rate analysis**

Taking the derivative of the population function provides the instantaneous rate of change. From this result we can see that the growth rate:

$$P(t) = 30e^{0.03t}$$


Remains positive at all times

Rises exponentially as time goes on

### **Figure interpretation**

The plotted curve displays a steadily rising exponential shape. Growth starts off gradually, then speeds up in the later years. The highest rate of increase appears near the end of the 20-year interval, suggesting that population pressure continues to intensify over the period.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a time range from 0 to 20 years
time_vals = np.linspace(0, 20, 200)

# Compute population using an exponential model
population_vals = 1000 * np.exp(0.03 * time_vals)

# Plot the curve
plt.figure()
plt.plot(time_vals, population_vals)
plt.xlabel("Years")
plt.ylabel("Number of People")
plt.title("Exponential Growth Curve for Paris Population")
plt.show()
```

```python
# Compute the rate of change of the population
growth_rate = 30 * np.exp(0.03 * time_vals)

# Create a new plot for the growth rate curve
plt.figure()
plt.plot(time_vals, growth_rate)
plt.xlabel("Years")
plt.ylabel("Rate of Increase")
plt.title("Change in Population Over Time")
plt.show()
```

# **Probability Analysis of Immigration Events Using a Poisson Distribution**
The yearly count of random immigration events is described using a Poisson distribution with a given rate parameter.

$$\lambda = 5$$


### **Expected value and variance**

For a Poisson model:

* The expected number of events (mean) is 5
* The variance is also 5

### **Figure interpretation**

The bar chart indicates that:

* The most common number of events falls roughly between 3 and 7 per year
* The highest probability occurs at 5 events
* Very low or very high counts have small probabilities and are therefore rare.


```python
from scipy.stats import poisson

# Possible event counts
events = np.arange(0, 15)

# Poisson probabilities with mean rate 5
pmf_values = poisson.pmf(events, 5)

# Display as a bar chart
plt.figure()
plt.bar(events, pmf_values)
plt.xlabel("Count of Immigration Occurrences")
plt.ylabel("Probability Mass")
plt.title("Poisson PMF with Mean = 5")
plt.show()
```

# **Statistical Analysis of District Population Data**
Population figures from eight Paris districts were examined to estimate the typical district size. The calculated sample average was **1064.38** residents, and the sample standard deviation was **12.081**, showing only minor differences among districts. A **95% confidence interval** for the true mean district population was found to be **[1054.27, 1074.48]**.

Overall, these findings indicate that district populations are fairly uniform. Even though the dataset is limited, the relatively tight confidence interval suggests the mean estimate is reasonably dependable. Such estimates are useful for tracking population patterns and guiding city resource allocation decisions.


```python
from scipy import stats

# Population figures for several districts
pop_values = np.array([1050, 1070, 1065, 1080, 1045, 1075, 1060, 1070])

# Number of observations
sample_count = len(pop_values)

# Compute average
avg_pop = np.mean(pop_values)

# Compute sample standard deviation (uses n-1)
sample_sd = np.std(pop_values, ddof=1)

# Build a 95% confidence interval with the t distribution
conf = 0.95
degrees_freedom = sample_count - 1
t_value = stats.t.ppf((1 + conf) / 2, degrees_freedom)

error_margin = t_value * (sample_sd / np.sqrt(sample_count))
lower_bound = avg_pop - error_margin
upper_bound = avg_pop + error_margin

# Display summary statistics
print("Mean:", round(avg_pop, 2))
print("Std Dev:", round(sample_sd, 3))
print("95% CI: [", round(lower_bound, 2), ",", round(upper_bound, 2), "]")

# Draw a horizontal boxplot and mark mean and interval limits
plt.boxplot(pop_values, vert=False)
plt.axvline(avg_pop)
plt.axvline(lower_bound, linestyle="--")
plt.axvline(upper_bound, linestyle="--")

plt.xlabel("Population Size")
plt.title("District Population Spread with Mean and 95% Confidence Interval")
plt.show()

```
