```python
from cryptography.fernet import Fernet

key = Fernet.generate_key()

cipher = Fernet(key)

name=input('Enter Users Name: ')
email=input('Enter Users Email: ')

record_name = name

encrypted_record_name = cipher.encrypt(record_name.encode())

print(encrypted_record_name)

record_email = email

encrypted_record_email = cipher.encrypt(record_email.encode())

print(encrypted_record_email)


decrypted_record_name = cipher.decrypt(encrypted_record_name).decode()
decrypted_record_email = cipher.decrypt(encrypted_record_email).decode()


print(decrypted_record_name)
print(decrypted_record_email)


```
